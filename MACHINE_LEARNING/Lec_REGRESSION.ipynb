```
# 1468 * 81짜리 데이터, 일일히 null값을 찾고 변경하기 어려우니깐 함수로 정의
def check_fill_na(data):
  new_df = data.copy()
  new_df_na = (new_df.isnull().sum() / len(new_df)) * 100
  new_df_na.sort_values(ascending=False).reset_index(drop=True)
  new_df_na = new_df_na.drop(new_df_na[new_df_na == 0].index).sort_values(ascending=False)
  return new_df_na

check_fill_na(df_train)
```

```
#결측치를 어떻게 처리할지 함수로 정의함
def fill_missing(df, cols, val):
    """ val 입력값을 넣는다. """
    for col in cols:
        df[col] = df[col].fillna(val)

def fill_missing_with_mode(df, cols):
    """ 최대 빈도수를 넣는다. """
    for col in cols:
        df[col] = df[col].fillna(df[col].mode()[0])
        
def addlogs(res, cols):
    """ 로그 변환 """
    m = res.shape[1]
    for c in cols:
        res = res.assign(newcol=pd.Series(np.log(1.01+res[c])).values)   
        res.columns.values[m] = c + '_log'
        m += 1
    return res
```

- 범주형 데이터 다루기

```
def fix_missing_cols(in_train, in_test):
    missing_cols = set(in_train.columns) - set(in_test.columns)
    # 테스트 데이터와 훈련 데이터의 컬럼을 동일하게 하는 코드는 작성한다. 
    for c in missing_cols:
        in_test[c] = 0
    # 순서를 동일하게 만든다. 
    in_test = in_test[in_train.columns]
    return in_test

def dummy_encode(in_df_train, in_df_test):
    df_train = in_df_train
    df_test = in_df_test
    categorical_feats = [
        f for f in df_train.columns if df_train[f].dtype == 'object'
    ]
    print(categorical_feats)
    for f_ in categorical_feats:
        prefix = f_
        df_train = pd.concat([df_train, pd.get_dummies(df_train[f_], prefix=prefix)], axis=1).drop(f_, axis=1)
        df_test = pd.concat([df_test, pd.get_dummies(df_test[f_], prefix=prefix)], axis=1).drop(f_, axis=1)
        df_test = fix_missing_cols(df_train, df_test)
    return df_train, df_test
```

```
#cross_val_score를 기준으로 RMSE를 도출할거야
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats
from sklearn.datasets import load_boston #boston은 iris와 마찬가지로 sklearn.datasets에서 제공해주는 데이터 중 하나이다
```

- 머신러닝 모형 만들기

```
from sklearn.model_selection import train_test_split #data분리용 패키지
from sklearn.linear_model import LinearRegression #선형회귀모형 패키지
from sklearn.metrics import mean_squared_error, r2_score #RMSE계산용 패키지_정확히 말하면 RMSE는 아니고 MSE, r2_score는 결정계수인 것 같다
```

```
#훈련데이터, 테스트데이터 분리
y_target = bostonDF['PRICE'] #종속변인인 가격
x_data = bostonDF.drop(['PRICE'], axis = 1, inplace=False) #종속변수인 price를 빼고 나머지를 독립변수로 넣어준거

x_train, x_test, y_train, y_test = train_test_split(x_data, y_target, test_size = 0.3, random_state=1)
#x_data는 독립변인, y_target은 종속변인인걸 기억하자
```

```
#선형회귀 모형 만들기
lr_model = LinearRegression()
lr_model.fit(x_train,y_train) #훈련용 데이터로 교육
```

```
#예측
y_preds = lr_model.predict(x_test) #테스트용 독립변인을 바탕으로 y(price)값을 predict
y_preds #예측한 가격 데이터
```

- MSE(Mean_squarred_error)로 RMSE를 구하기

  - 사용이유 = sklearn은 RMSE를 제공하지 않기 때문
  - MSE에 루트를 씌우면 RMSE가 된다

    - MSE를 그냥 두면 제곱 값이기 때문에, 실제 오차보다 더 커지는 경향성이 있다




```
mse = mean_squared_error(y_test, y_preds)
print(mse)
rmse = np.sqrt(mse) #np.sqrt는 np에 들어있는 루트함수
print(rmse) #이렇게 도출한 rmse가 어떤 정도인지 모르기 때문에 비교를 위해 k폴드 교차검증 시행
```

```
# K폴드 교차검증 by cross_val_score()
from sklearn.model_selection import cross_val_score
neg_mse_scores = cross_val_score(lr_model, x_data, y_target, scoring='neg_mean_squared_error', cv=5) #cv는 k폴드검증의 횟수 #neg_mean_squarred_error는 제곱의 평균을 구하되, 값을 음수로 반환
rmse_scores=np.sqrt(-1 * neg_mse_scores) #앞에서 반환값이 음수기 때문에 -1을 곱해주어야한다
print(rmse_scores)
```

```
avg_rmse = np.mean(rmse_scores)
print(avg_rmse) #k폴드값의 평균이 전체 rmse값이라고 생각할 수 있다
```

  - 앞에서 drop하는 변수를 추가해서 적어주면 된다
